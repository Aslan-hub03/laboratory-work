```python
import math


# 1. Класс для ввода и вывода строки в верхнем регистре
class StringManipulator:
    def getString(self):
        self.text = input("Введите строку: ")

    def printString(self):
        print(self.text.upper())


# 2. Классы Shape и Square с вычислением площади
class Shape:
    def area(self):
        return 0


class Square(Shape):
    def __init__(self, length):
        self.length = length

    def area(self):
        return self.length ** 2


# 3. Класс Rectangle, наследующий от Shape
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width


# 4. Класс Point с методами для отображения, перемещения и вычисления расстояния между точками
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def show(self):
        print(f"Координаты точки: ({self.x}, {self.y})")

    def move(self, x, y):
        self.x = x
        self.y = y

    def dist(self, other_point):
        return math.sqrt((self.x - other_point.x) ** 2 + (self.y - other_point.y) ** 2)


# 5. Класс BankAccount с методами deposit и withdraw
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} добавлено. Баланс: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Недостаточно средств!")
        else:
            self.balance -= amount
            print(f"{amount} снято. Баланс: {self.balance}")


# 6. Фильтрация простых чисел с использованием filter и lambda
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


def main():
    # Тест задачи 1
    print("\nЗадача 1:")
    string_obj = StringManipulator()
    string_obj.getString()
    string_obj.printString()

    # Тест задачи 2
    print("\nЗадача 2:")
    square = Square(5)
    print(f"Площадь квадрата со стороной 5: {square.area()}")

    # Тест задачи 3
    print("\nЗадача 3:")
    rectangle = Rectangle(4, 5)
    print(f"Площадь прямоугольника 4x5: {rectangle.area()}")

    # Тест задачи 4
    print("\nЗадача 4:")
    p1 = Point(0, 0)
    p2 = Point(3, 4)
    p1.show()
    p2.show()
    print(f"Расстояние между точками: {p1.dist(p2)}")

    # Тест задачи 5
    print("\nЗадача 5:")
    account = BankAccount("Иван", 100)
    account.deposit(50)
    account.withdraw(30)
    account.withdraw(200)

    # Тест задачи 6
    print("\nЗадача 6:")
    numbers = [1, 2, 3, 4, 5, 10, 11, 13, 15]
    prime_numbers = list(filter(lambda x: is_prime(x), numbers))
    print(f"Простые числа: {prime_numbers}")


if __name__ == "__main__":
    main()
```
